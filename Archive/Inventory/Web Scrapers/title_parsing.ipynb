### Extraxt Useful Queries from the List of titles. (Keywords for Search)

```python
from collections import Counter
```

```python
with open('titles.txt','r') as f:
    data = f.readlines()
```

```python
a1,a2,a3 = data
```

```python
a1 = dict(Counter(' '.join(a1.split(';;')).split(' ')))
a2 = dict(Counter(' '.join(a2.split(';;')).split(' ')))
a3 = dict(Counter(' '.join(a3.split(';;')).split(' ')))
```

```python
s1 = set(a1.keys())
s2 = set(a2.keys())
s3 = set(a3.keys())
```

```python
hand_selected = []
```

```python
complet_set = s1 & s2 & s3
complet_set
```

```python
for value in s1:
    print(f'item: {value}')
    if input('Enter') == '1':
        hand_selected.append(value)
```

```python
hand_selected
```

```python
with open(f'queries.txt','w') as f:
    for t in hand_selected:
        f.write(t + '\n')
```

```python
hand_selected = list(set(hand_selected))
```
