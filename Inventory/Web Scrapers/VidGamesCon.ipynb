```python
from bs4 import BeautifulSoup
import requests
```

```python
items = requests.get(url = 'https://videogamecons.com/calendar/calendar.php?year=2025')
soup  = BeautifulSoup(items.content,'html.parser')
webpage_text = soup.get_text()
```

```python
items = webpage_text.split('\n')
```

```python
items = [i.strip() for i in items if i != '']
```

```python
items
```

```python
start_idx = items.index('Location')
items    = items[start_idx + 1:]
end_idx  =  items.index('Convention')
events = items[: end_idx]
```

```python
for name, date, loc in list(zip(*[iter(events)]*3)):
    print(name,date,loc,sep="|")
```
