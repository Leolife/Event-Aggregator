```python
from bs4 import BeautifulSoup
import requests
```

```python
items = requests.get(url = 'https://animecons.com/events/')
soup  = BeautifulSoup(items.content,'html.parser')
webpage_text = soup.get_text()
```

```python
items = webpage_text.split('\n')
```

```python
items = [i.strip() for i in items if i != '']
idx = items.index('Location')
end_idx = items[idx + 1:].index('Dates')
items = items[idx + 1: end_idx]
assert(len(items) % 3 == 0)
```

```python
for name, date, loc in list(zip(*[iter(items)]*3)):
    print(name,date,loc,sep="|")
```
